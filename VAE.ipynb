### А теперь VAE

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
BATCH_SIZE = 250
LATENT = 4
```

```python
class VAE(nn.Module):
    def __init__(self):
        super(VAE, self).__init__()
        
        self.el1 = nn.Linear(in_features=28*28, out_features=256)
        self.b1 = nn.BatchNorm1d(256)
        self.el2 = nn.Linear(in_features=256, out_features=128)
        self.b2 = nn.BatchNorm1d(128)
        self.meanF = nn.Linear(in_features=128, out_features=LATENT)
        self.varF = nn.Linear(in_features=128, out_features=LATENT)
        
        self.dl1 = nn.Linear(in_features=LATENT, out_features=128)
        self.b3 = nn.BatchNorm1d(128)
        self.dl2 = nn.Linear(in_features=128, out_features=256)
        self.b4 = nn.BatchNorm1d(256)
        self.dl3 = nn.Linear(in_features=256, out_features=28*28)
        
        
        self.drop = nn.Dropout(p=0.01)
        
    def forward(self, x):
        el1 = F.relu(self.el1(x))
        b1 = self.drop(self.b1(el1))
        el2 = F.relu(self.el2(b1))
        b2 = self.drop(self.b2(el2))
        
        self.mean = self.meanF(b2)
        self.var = self.varF(b2)
        
        epsilon = torch.normal(mean=0.0, std=1.0, size=(x.shape[0], LATENT), device=x.device)
        self.z = self.mean + torch.exp(self.var / 2.0) * epsilon
        
        dl1 = F.relu(self.dl1(self.z))
        b3 = self.drop(self.b3(dl1))
        dl2 = F.relu(self.dl2(b3))
        b4 = self.drop(self.b4(dl2))
        dl3 = torch.sigmoid(self.dl3(b4))
        
        return dl3
    
    def decode(self, z):
        dl1 = F.relu(self.dl1(z))
        b3 = self.drop(self.b3(dl1))
        dl2 = F.relu(self.dl2(b3))
        b4 = self.drop(self.b4(dl2))
        dl3 = torch.sigmoid(self.dl3(b4))
        
        return dl3
        
vae = VAE().cuda()
```

```python
BCELossF = nn.BCELoss()
def vae_loss(x, mean, var, target):
    x = x.view(-1, 28*28)
    target = target.view(-1, 28*28)
    xent_loss = 28*28*BCELossF(x, target)
    kl_loss = -0.5 * torch.sum(1 + var - torch.square(mean) - torch.exp(var))
    
#     return (xent_loss + kl_loss)/2/28/28
    return xent_loss/(28*28)
# vae_loss(torch.rand(1, 1, 28, 28), 1, 1, torch.rand(1, 1, 28, 28))
```

class VAE_c(nn.Module):
    def __init__(self):
        super(VAE_c, self).__init__()
        
        self.ec1 = nn.Conv2d(in_channels=1, out_channels=4, kernel_size=3, padding=1)
        self.ec2 = nn.Conv2d(in_channels=4, out_channels=2, kernel_size=3, padding=1, stride=2)
        self.ec3 = nn.Conv2d(in_channels=2, out_channels=1, kernel_size=3, padding=1)
        self.ec4 = nn.Conv2d(in_channels=1, out_channels=1, kernel_size=3, padding=1, stride=2)

        self.mean = nn.Linear(in_features=49, out_features=2)
        self.var = nn.Linear(in_features=49, out_features=2)
        
        self.dl1 = nn.Linear(in_features=2, out_features=49)
        self.dc1 = nn.Conv2d(in_channels=1, out_channels=2, kernel_size=3, padding=1)
        self.dc2 = nn.ConvTranspose2d(in_channels=2, out_channels=2, kernel_size=2, padding=0, stride=2)
        self.dc3 = nn.Conv2d(in_channels=2, out_channels=4, padding=1, kernel_size=3)
        self.dc4 = nn.ConvTranspose2d(in_channels=4, out_channels=1, kernel_size=2, padding=0, stride=2)
        
        
    def forward(self, x):
        x = F.relu(self.ec1(x))
        x = F.relu(self.ec2(x))
        x = F.relu(self.ec3(x))
        x = F.relu(self.ec4(x))
        x = x.view(-1, 49)
        
        mean = self.mean(x)
        var = self.var(x)
        
        epsilon = torch.normal(mean=0.0, std=1.0, size=(x.shape[0], 2), device=x.device)
        z = mean + torch.exp(var / 2.0) * epsilon
        self.z = z
        
        x = F.relu(self.dl1(z))
        x = x.view(-1, 1, 7, 7)
        x = F.relu(self.dc1(x))
        x = F.relu(self.dc2(x))
        x = F.relu(self.dc3(x))
        x = torch.sigmoid(self.dc4(x))
        
        return x
    
    def decode(self, z):
        x = F.relu(self.dl1(z))
        x = x.view(-1, 1, 7, 7)
        x = F.relu(self.dc1(x))
        x = F.relu(self.dc2(x))
        x = F.relu(self.dc3(x))
        x = torch.sigmoid(self.dc4(x))
        
        return x
        
VAE_c = VAE_c().cuda()

```python
import torchvision
import torchvision.transforms as transforms

transform = transforms.Compose(
    [transforms.ToTensor()])

trainset = torchvision.datasets.MNIST(root='.', train=True, download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=BATCH_SIZE, shuffle=True)

testset = torchvision.datasets.MNIST(root='.', train=False, download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=BATCH_SIZE*4, shuffle=False)
```

```python
losses = []
```

```python
import torch.optim as optim

loosFunction = nn.BCELoss()
optimizer = optim.Adam(vae.parameters(), lr=0.0001)
```

```python
for epoch in range(10):  # loop over the dataset multiple times

    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        inputs = data[0].cuda()

        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = vae(inputs.view(-1, 28*28))
#         loss = vae_loss(outputs, vae.mean, vae.var, inputs.view(-1, 28*28))
        loss = loosFunction(outputs, inputs.view(-1, 28*28))
        loss.backward()
        optimizer.step()

        # print statistics
        running_loss += loss.item()
        
    print('[%d, %5d] loss: %.5f' %
          (epoch + 1, i + 1, running_loss / len(trainloader)))
    losses.append(running_loss / len(trainloader))
    running_loss = 0.0
    
plt.plot(losses)
plt.show()
```
